## (c) 
Code up the Gradient Decent algorithm from your pseudocode in actual Python [you should use numpy expressions] using a suitable starting guess.

```python
# The initial function, y = -exp(-(x - 3.14) ** 2).
def f_x(x):
    f = -np.exp(-(x - 3.14) ** 2)
    return f

# The gradient of initial function, y = -exp(-(x - 3.14) ** 2).
def gradient(x):
    gradient = 2 * (x - 3.14) * np.exp(-(x - 3.14) ** 2)
    return gradient

# Set the scalar multiplication factor eta.
eta = 0.001

# Set the threshold of iteration times, N_GD_iterations.
N_GD_iterations = 500000

# Set initial guess at the solution x_t.
x_t = 6

for t in range(N_GD_iterations):
    if (np.abs(gradient(x_t)) > 0.00001): 
        x_t = x_t - eta * gradient(x_t)
    else:    
        break

print("The solution is " + "%f"%x_t + ".")

result = f_x(x_t)
print("The minimum of the objective function is " + "%f"%result + ".")
```

If the problem would have been instead to consider the function $f(x) = e^{-(x-3.14)^2}$ and were asked to find the maximum locating and value: How would you solve this and what part of the Gradient Descent algorithm would you modify? There are two ways of doing this that both boil down to the same thing.

The first method by modifying some parts of the previous codes for the minization problem:
1. Code up the function of interest we want to maximize.
2. Code up the gradient of the function we want to maximize.
3. Code up an iterative loop with the update equation:
     $x_t = x_t + \eta_t * gradient(x_t)$


```python
# The initial function, y = exp(-(x - 3.14) ** 2).
def f_x(x):
    f = np.exp(-(x - 3.14) ** 2)
    return f

# The gradient of initial function y = exp(-(x - 3.14) ** 2).
def gradient(x):
    gradient = -2 * (x - 3.14) * np.exp(-(x - 3.14) ** 2)
    return gradient

# Set the scalar multiplication factor eta_t.
eta = 0.01

# Set the threshold of iteration times, N_GD_iterations.
N_GD_iterations = 500000

# Set initial guess at the solution x_t.
x_t = 6

for t in range(N_GD_iterations):
    if (np.abs(gradient(x_t)) > 0.0001):
        x_t = x_t + eta * gradient(x_t)
    else:
        break
    
print("The solution is " + "%f"%x_t + ".")

result = f_x(x_t)
print("The maximum of the objective function is " + "%f"%result + ".")
```

Formulate the above problem as a simple optimization problem in one line of text using

$V = \pi * r^2 * h = 355$, where $V$ is the volume of the can, $h$ is the height of the can, and $r$ is the radius of the bottom of the can.

First, we could plot the $s-r$ function to observe how the curve looks like.

```python
# The objective function
def cansurface(r):
    s = 710/r + 2 * math.pi * (r ** 2)
    return s

# The radius of the bottom.
r = np.arange(1, 15, 0.1)
# The area of the surface.
s = cansurface(r)

plt.title("The surface area of the can") 
plt.xlabel("radius") 
plt.ylabel("s") 
plt.plot(r,s) 
plt.show()
```

From the plot above, we can see that the minimum of the surface area is located in $x$ within the range from $x = 0$ to $x = 6$. 
<br>
We can do the same thing as before to work out the suitable value of radius and its minimal surface area.

```python
# The gradient of the objective function.
def gradient(r):
    gradient = 4 * math.pi * r - (710 / (r**2))
    return gradient

# Set the scalar multiplication factor eta_t.
eta_t = 0.001

# Set the threshold of iteration times, N_GD_iterations.
N_GD_iterations = 500000

# Set initial guess at the solution r_0, let's say 7.
r_t = 7

for t in range(N_GD_iterations):
    if (np.abs(gradient(r_t)) > 0.001):
        r_t = r_t - eta_t * gradient(r_t)
    else:
        break

print("The radius at the bottom, r, is " + "%f"%r_t + ".")

A = cansurface(r_t)
print("The surface area is " + "%f"%A + ".")
```
