3. In Eq (8), we have our predicion model. This is a mathematical function describing what we think is the true relationship between the quantities $t$ (time traveled) and $s$ (distance traveled) is. Our model contains one parameter $v$ that we would like to and the true value of. First, let us assume $v = 0$: What would the resulting model predict for each of the four values tn in our dataset D? What predictions would we get if we instead use $v = 35$ or
$v = 42$? For each of the three parameter values, plot their the corresponding lines in the same graph together.

```python
# Plot the dataset points
plt.scatter(D[:, 0], D[:, 1])

# When v = 0
v = 0
t = np.arange(0, 10, 0.02)
s = v * t
plt.text(7, 10, s = "v = 0")
plt.plot(t, s)

# When v = 35
v = 35
t = np.arange(0, 10, 0.02)
s = v * t
plt.text(7, 220, s = "v = 35")
plt.plot(t, s)

# When v = 42
v = 42
t = np.arange(0, 10, 0.02)
s = v * t
plt.xlabel("t")
plt.ylabel("s")
plt.text(7, 300, s = "v = 42")
plt.plot(t, s)
```
